```python
# To print array of m*n conisiting of zeroes only

np.zeros((3, 4))  #((m*n)) filled with zeros
```

```python
# To print array of m*n consisting of ones only

np.ones((3,4))   #((m*n)) filled with ones
```

```python
# To print array of m*n conisiting of zeroes only

np.full((3, 4),5)    #((m*n),value to be printed)
```

```python
# To print array of m*m identity matrix

np.eye(3)
```

```python
# To print array of like arange(from, to, by adding the no)

np.arange(1, 100, 30)
```

```python
# To print array with evenly spaced

np.linspace(1, 100, 10)
```
